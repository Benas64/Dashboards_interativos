```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
petr4=pd.read_csv('Petr4.SA.csv')
petr4.head()
```

```python
#Vamos ver o tamanho do dataframe
petr4.shape
```

```python
#Vamos ver as colunas do dataframe
petr4.columns
```

```python
#Agora vamos fazer um gráfico de linha com a coluna Close
petr4['Close'].plot(figsize=(10,5))
```

```python
#Agora vamos fazer um gráfico de linha com a coluna Open
petr4['Open'].plot(figsize=(10,5))
```

```python
#Agora vamos fazer um boxplot com a coluna Close
petr4['Close'].plot(kind='box',figsize=(10,5))
```

```python
#Agora vamos fazer um boxplot com a coluna Open e a close
petr4[['Open','Close']].plot(kind='box',figsize=(10,5))
```

```python
#Agora vamos fazer um histograma com a coluna Close e a open 
petr4[['Open','Close']].plot(kind='hist',figsize=(10,5))
```

```python
#Agora vamos fazer umboxplot com as colunas open e close ao longo dos anos usando seaborn
sns.boxplot(data=petr4[['Open','Close']],x=petr4['Date'].str[:4],y='Open')
```

```python
#vamosfazer um gráfico de linha com as colunas open e close ao longo dos anos usando seaborn
sns.lineplot(data=petr4[['Open','Close']],x=petr4['Date'].str[:4],y='Open')
```

```python
#Vamos fazer um grafico boxplot com o menor valor e o maior lado a lado
sns.boxplot(data=petr4[['High']],x=petr4['Date'].str[:4],y='High', color='green')



```

```python
sns.boxplot(data=petr4[['Low']],x=petr4['Date'].str[:4],y='Low', color='red')
```

```python
sns.scatterplot(data=petr4[['Open','Close']],x='Open',y='Close')
```

```python
#Agora vamos fazer um gráfico de linha com a coluna Close
petr4['Close'].plot(figsize=(10,5),color='red')
```

```python
#Agora vamos fazer um gráfico de linha com a coluna Open e depois juntar com o gráfico de linha da coluna Close
petr4['Open'].plot(figsize=(10,5),color='blue',x='date')
petr4['Close'].plot(figsize=(10,5),color='red',x='date')

```

```python
#Agora vamos fazer um grafic boxblot com o volume ao lono dos anos
sns.boxplot(data=petr4[['Volume']],x=petr4['Date'].str[:4],y='Volume', color='green')


```

```python
#aqui vamos fazer um gráfico de linha com o volume ao longo dos anos
sns.lineplot(data=petr4[['Volume']],x=petr4['Date'].str[:4],y='Volume')
```
